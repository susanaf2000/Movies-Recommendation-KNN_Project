```python
# Your code here
import pandas as pd

data_movies = pd.read_csv('../data/raw/tmdb_5000_movies.csv')
data_credits = pd.read_csv('../data/raw/tmdb_5000_credits.csv')

data_movies.head()
```

```python
data_credits.head()
```

```python
import sqlite3

con = sqlite3.connect('../data/raw/database_movies.db')

data_movies.to_sql("movies", con, if_exists = "replace", index = False)
data_credits.to_sql("credits", con, if_exists = "replace", index = False)
```

```python
query = """
    SELECT *
    FROM movies
    INNER JOIN credits
    ON movies.title = credits.title;
"""
```

Now that we have our tables together by title, let's create a third table with this info:

```python
data = pd.read_sql(query, con)
con.close()
```

```python
data.head()
```

```python
data.columns
```

```python
data = data[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
data.head()
```

```python
data = data.loc[:, ~data.columns.duplicated()]
```

```python
data.head()
```

As we can see, the genres, keywords and cast columns are JSON formatted. I'll select only the name attribute for the keywords and genres and the first three names for the cast:

```python
import json

def load_json_safe(json_str, default_value=None):
    try:
        return json.loads(json_str)
    except (TypeError, json.JSONDecodeError):
        return default_value

data['genres'] = data['genres'].apply(lambda x: [i['name'] for i in load_json_safe(x, [])])
data['keywords'] = data['keywords'].apply(lambda x: [i['name'] for i in load_json_safe(x, [])])
data['cast'] = data['cast'].apply(lambda x: [i['name'] for i in load_json_safe(x, [])[:3]])

# Extract director from crew column
data['crew'] = data['crew'].apply(lambda x: " ".join([crew_member['name'] for crew_member in load_json_safe(x, []) if crew_member.get('job') == 'Director']))

# Convert overview column to string
data["overview"] = data["overview"].astype(str)

# Combine all columns into a single tags column
data["tags"] = data["overview"] + " " + data["genres"].apply(lambda x: " ".join(x)) + " " + data["keywords"].apply(lambda x: " ".join(x)) + " " + data["cast"].apply(lambda x: " ".join(x)) + " " + data["crew"]


data["tags"] = data["tags"].str.replace(",", " ")

# Drop unnecessary columns
data.drop(columns=["genres", "keywords", "cast", "crew", "overview"], inplace=True)

```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.neighbors import NearestNeighbors

vec_model = CountVectorizer()
data_vec = vec_model.fit_transform(data['tags'])
```

```python
model = NearestNeighbors(n_neighbors = 6, algorithm = "brute", metric = "cosine")
model.fit(data_vec)
```

```python
def recommend(movie):
    movie_index = data[data["title"] == movie].index[0]
    distances, indices = model.kneighbors(data_vec[movie_index], n_neighbors=6)
    
    print(f"Top 5 recommendations for '{movie}':")
    for i in range(1, len(distances[0])):
        idx = indices[0][i]
        print(data.iloc[idx]['title'])

recommend("Dune")
```
